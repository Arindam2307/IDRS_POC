```python
filter=df['SEND_RECORD'].values>1500000
df_outlier_rem=df[filter]
df_outlier_rem
```

```python
df_outlier_rem.shape
```

```python
df_outlier_rem.SEND_RECORD_NEW.plot.line(y='SEND_RECORD_NEW')
plt.title('Send record trend for values more than 150000')
plt.xlabel('INDEX')
plt.ylabel('SEND_RECORD')
plt.savefig('SEND_RECORD_Trend.png')
## Here we can reset the index and plot only 56 values starting from 0 to 55
```

```python
df_outlier_rem.SEND_RECORD.plot.bar(y='SEND_RECORD')
plt.title('Send record trend for values more than 150000')
plt.xlabel('INDEX')
plt.ylabel('SEND_RECORD')
```

```python
plt.hist(df_outlier_rem['SEND_RECORD_NEW'], histtype='bar')
plt.title('Send record trend for values more than 150000')
plt.xlabel('INDEX')
plt.ylabel('SEND_RECORD')
```

```python
features_to_normalize = ['SEND_RECORD']
df_outlier_rem[features_to_normalize] = df_outlier_rem[features_to_normalize].apply(lambda x:(x-x.min()) / (x.max()-x.min()))
df_outlier_rem
```

```python
df_outlier_rem["Variation"]=df_outlier_rem["SEND_RECORD"]*100
df_outlier_rem
```

```python
df_outlier_rem.drop(["TRANS_STATUS","SEND_DATE_UTC","SEND_PRODUCT","REC_PRODUCT","RECEIVE_DATE_UTC"], axis=1,inplace=True)
```

```python
df_outlier_rem
```

```python
df_outlier_rem['Status'] = df_outlier_rem['Variation'].apply(lambda x: 1 if x > 80 else 0)
# Here we have put status as an indicator-> 0 is Safe 1 is Alert and it would indicate record instances which have exceeded 
#1700000 barrier
```

```python
#df_outlier_rem.reset_index(inplace=true) can be done to reset the index to 0
df_outlier_rem
```

```python
df_outlier_rem
```

```python
df_outlier_rem.dtypes
```

```python
df_outlier_rem["Variation_Ceil"]=df_outlier_rem["Variation"].apply(np.ceil)
df_outlier_rem
```

```python
df_outlier_rem.dtypes
```

```python
df_outlier_rem['Variation_Ceil']=df_outlier_rem['Variation_Ceil'].astype(int)
```

```python
df_outlier_rem.dtypes
```

```python
df_outlier_rem
```

```python
df_outlier_rem.Variation_Ceil.plot.bar(x='index', y='Variation_Ceil')
plt.title('Values close to the threshold limit')
plt.xlabel('index')
plt.ylabel('Variation_Ceil')
plt.savefig('Variation_Trend_Graph.png')
## This figure exhibits the graph of VARIATION_CEIL after normalization for send record values greater than 150000. 
## frequently we observe the trend reaching close and as well as exceeding the threshold barrier
```

```python
df_outlier_rem.Status.plot.bar(x='index')
plt.title('Values close to the threshold limit')
plt.xlabel('index')
#plt.ylabel('Variation_Ceil')
plt.savefig('Status_Graph.png')
## This figure exhibits the Status graph after normalization where there is a significant chance of job failing for exceeding IDRS limit. 
```
